# 階層ベイズモデル検討(ちょっと雑です．．．)

## 空間構造のある階層ベイズモデル

- 緑本第11章で，階層ベイズモデルの応用例として，空間構造のあるベイズ統計モデルが紹介されていた．
- この手法は「データをとる場所の空間配置の影響」を考慮したベイズ統計モデルで，本内では場所差の空間相関を考慮した統計モデリングを紹介している

## 11章の空間統計モデル

- 一次元的に分布する50個の調査区画{j}それぞれで架空植物の個体数{yj}を数える
- 区画{j}ごとに平均{λj}が異なるポアソン分布にしたがうと考える．
$$
{p(y_j | \lambda_j)
= \frac{\lambda_j^{y_j} \exp (-\lambda_j)}{y_j!}
}
$$
- 平均個体数{λj}は線形予測子と対数リンク関数を使って以下のように表す
$$
{\log \lambda_j = \beta + r_j
}
$$
    - {β}：大域的なパラメータ．すべての場所に同じ影響を与える． 無情報事前分布(平均0,標準偏差100ぐらいの平べったい正規分布)
    - {rj}：局所的なパラメータ．区画{j}だけに影響を与える． 階層事前分布(平均0(空間相関を考慮するなら変わる),標準偏差{s}に従う正規分布)


- 第10章では以下の階層事前分布を指定
$$
{p(r_j|s) = \frac{1}{\sqrt{2 \pi s^2}} \exp \left( - \frac{r_j^2}{2 s^2} \right)
}
$$
- こう指定すると，場所差{rj}がそれぞれ独立に同じ事前分布を仮定していることになるので，空間相関を考慮できてない

- なので，事前分布に工夫を加える(仮定を設ける)

     - 「近傍」にしか区間の場所差は影響されない
     - 「近傍」の数は有限，区間はモデル設計者が決める
     - 「近傍」の直接の影響は等しい


- {rj}の条件付き事前分布が以下のような正規分布であると設定（条件つき自己回帰モデル，conditional auto regressive, CAR）
$$
{p(r_j|\mu_j, s) = \frac{n_j}{\sqrt{2 \pi s^2}} \exp \left( - \frac{(r_j-\mu_j)^2}{2 s^2/n_j} \right)
}
$$
- ここでの正規分布の平均が，「近傍」の平均値に等しいとする
$$
{\mu_j = \frac{r_{j-1} + r_{j+1}}{2}
}
$$
- 標準偏差は
$$
s/\sqrt{n_j}
$$

- ばらつきパラメータ{s}はどの場所でも同じと仮定(連続一様分布の無情報事前分布を指定する）

- 事後分布は
$$
{p(\beta, s, \{r_j\} | Y)
\propto p(\{r_j\} | s) p(s) p(\beta) \prod_j p(y_j | \lambda_j)
}
$$

- その後，MCMCサンプリングで，各パラメータの事後分布を推定する
